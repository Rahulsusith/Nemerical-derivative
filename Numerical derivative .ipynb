# Nemerical Derivative 


## Here I am going to do Nemerical derivative of a function using Matplotlib and numpy


```python
##importing numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt

```

```python
## lets take an equation y=3xÂ²+2x+1
```

```python
def equ(x):
    return 3*x**2+2*x+1
```

```python
## let create x values of linear space between 0 to 10 of 10 values 
x=np.linspace(0,10,10)
print(x)
```

```python
## y values

y=equ(x)
print(y)
```

```python
## lets plot the equation 

plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.show()
```

```python
def derivative(x,y):
    
    ##derivative of y is dy/dx
    yd=np.diff(y)/np.diff(x)
    
    xd=np.array([])
    for i in range(len(yd)):
        t=(x[i]+x[i+1])/2
        xd=np.append(xd,t)
    return xd,yd
```

```python
xd,yd=derivative(x,y)
plt.figure(figsize=(8,6))
plt.plot(xd,yd,label='1st Derivative')
plt.legend()
plt.show()
```

```python
##Second derivative 
xd2,yd2=derivative(xd,yd)
plt.plot(xd2,yd2,label='2nd derivative')

## for the second derivative it becomes constant
```

```python
# final plot
plt.plot(x,y,label='Function')
plt.plot(xd,yd,label='1st derivative')
plt.plot(xd2,yd2,label='2nd derivative')
plt.legend()
plt.show()
```
